```python
import yfinance as yf
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, precision_score

import datetime as dt
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
import time
```

```python
#This analyses the accuracy analyst recommendations, data compares -2m and -3m recommendations, backtest month should be last month to the start of this month, this meanning current month

def rec_analysis(stock, k=0.5, compare = False):
  recs = yf.Ticker(stock).recommendations
  recs = recs.iloc[2:, 1:]

  old_bullarity = (recs.iloc[1,0] + recs.iloc[1,1])/recs.iloc[1].sum()
  new_bullarity = (recs.iloc[0,0] + recs.iloc[0,1])/recs.iloc[0].sum()

  if compare:
    if new_bullarity > old_bullarity:
      return 1
    else:
      return 0

  if new_bullarity > k:
    return 1
  else:
    return 0
```

```python
def change_days(date_str, days):
    import datetime as dt
    date = dt.datetime.strptime(date_str, "%Y-%m-%d")

    new_date = date + timedelta(days=days)

    return new_date.strftime("%Y-%m-%d")

def change_months(date_str, monthss):
  import datetime as dt
  date_obj = datetime.strptime(date_str, "%Y-%m-%d")

  new_date = date_obj + relativedelta(months=monthss)

  return new_date.strftime("%Y-%m-%d")

def backtest(stock, date):#function will return how much the stock price changed in the coming month
  enddate = change_months(date, 1)

  initial_price = yf.Ticker(stock).history(start = change_days(date, -5), end = date).tail(1) #the closing price before date
  new_price = yf.Ticker(stock).history(start = date, end = enddate).tail(1)#Closing price of final day

  #print(order)
  pct_change = (float(new_price["Close"].iloc[0])-float(initial_price["Close"].iloc[0]))/float(initial_price["Close"].iloc[0])

  print('\n',f"[------=====в—јпёЏв—јпёЏв—јпёЏв—јпёЏв—јпёЏ Backtesting {stock} [{date} --> {enddate}], {pct_change} в—јпёЏв—јпёЏв—јпёЏв—јпёЏв—јпёЏ=====--------]")
  return pct_change
```
